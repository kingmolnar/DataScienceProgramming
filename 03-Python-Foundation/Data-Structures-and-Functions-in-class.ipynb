```python
mult_less = []
mult_great = []

for x in range(1,5):
    for y in range(1,5):
        if x<=y:
            mult_less.append((x,y, x*y))
        else:
            mult_great.append((x,y, 0))
print(mult_less)
print(mult_great)
```

```python
def mtable_func():
    for x in range(1,5):
        for y in range(1,5):
            res = (x,y, x*y) if x<=y else (x,y, 0)
            return res

            
mtable_func()
```

```python
def mtable_gen(n):
    i = 0
    for x in range(1,n):
        for y in range(1,n):
            res = (x,y, x*y) if x<=y else (x,y, 0)
            print(f"--- {i} ---")
            i += 1
            yield res
```

```python
mtable_gen_list = list(mtable_gen(5))
```

## Functions, Processing data (def, lambda, map, filter)

```python
def foo_xy(t):
    x, y = t
    if x<=y:
        return (x, y, x*y)
    return 0
```

```python
def xy_gen(n):
    return ( (x,y) for x in range(1,n) for y in range(1,n) )
```

```python
list(xy_gen(5))
```

```python
list(map(foo_xy, xy_gen(5)))
```

```python
foo_l = lambda t: (t[0], t[1], t[0]*t[1])
```

```python
foo_l((2,4))
```

```python
list(map(lambda t: (t[0], t[1], t[0]*t[1]), xy_gen(5)))
```
